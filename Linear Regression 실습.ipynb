```python
import tensorflow as tf

tf.compat.v1.disable_eager_execution()

x_data = [[1,1] , [2,2], [3,3]]
y_data = [[10], [20], [30]]

X = tf.compat.v1.placeholder(tf.float32, shape=[None , 2])
Y = tf.compat.v1.placeholder(tf.float32, shape=[None , 1])

W = tf.Variable(tf.random.normal(shape= (2,1)), name = 'W')
b = tf.Variable(tf.random.normal(shape= (1,)), name = 'b')
```

```python
hypothesis = tf.matmul(X, W) + b  ## matrix multiplication
cost = tf.reduce_mean(tf.square(hypothesis - Y))
optimizer = tf.compat.v1.train.GradientDescentOptimizer(learning_rate = 0.01).minimize(cost)
```

```python
with tf.compat.v1.Session() as sess:
    sess.run(tf.compat.v1.global_variables_initializer())

    for step in range(50):
        c, W_, b_, _ = sess.run([cost, W, b , optimizer], feed_dict = {X : x_data, Y: y_data})
        print('Step: %2d\t loss: %.2f\t' % (step,c))

    print(sess.run(hypothesis, feed_dict = {X : [[4,4]]}))

```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD

x_data = np.array([[1], [2], [3]])
y_data = np.array([[10], [20], [30]])

model = Sequential([
    Dense(1)
])


model.compile(loss = 'mean_squared_error' , optimizer = SGD(learning_rate=0.1))

model.fit(x_data, y_data, epochs = 100) # epochs 복수형으로 쓰기
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers.legacy import Adam, SGD

x_data = np.array([[1], [2], [3]])
y_data = np.array([[10], [20], [30]])

model = Sequential([
  Dense(1)
])

model.compile(loss='mean_squared_error', optimizer=SGD(learning_rate=0.1))

model.fit(x_data, y_data, epochs=100) # epochs 복수형으로 쓰기!
```

```python
import numpy as np
from tensorflow import keras

# Prepare data
x_data = np.array([[1], [2], [3]])
y_data = np.array([[10], [20], [30]])

# Create the model using functional API
model = keras.Sequential([
    keras.layers.Dense(1, activation='linear'),  # Linear activation for linear regression
])

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_squared_error'])

# Train the model
model.fit(x_data, y_data, epochs=100, verbose=2)

# Evaluate the model
model_loss, model_mse = model.evaluate(x_data, y_data)
print(f"Model loss: {model_loss}")
print(f"Mean squared error: {model_mse}")

# Make predictions
new_data = np.array([[4], [5]])
predictions = model.predict(new_data)
print(f"Predictions for new data: {predictions}")
```
